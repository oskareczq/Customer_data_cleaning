### Data Cleaning and Transformation Workflow

In this notebook, we are working with a customer dataset stored in a pandas DataFrame named `df`. The dataset contains information about customers, including their contact details, address, and preferences. Below is a summary of the data cleaning and transformation steps performed so far:

1. **Importing Necessary Libraries**: 
    - We imported `pandas`, `numpy`, and `datetime` for data manipulation and analysis.
    - Display settings were adjusted to show all columns in the DataFrame.

2. **Loading the Dataset**:
    - The dataset was loaded from an Excel file named `Customer Call List.xlsx`.

3. **Data Cleaning**:
    - **Removing Duplicates**: Duplicate rows were dropped to ensure data integrity.
    - **Dropping Irrelevant Columns**: Columns deemed not useful for analysis were removed.
    - **Standardizing Text Data**: The `Last_Name` column was stripped of unwanted characters (e.g., `/`, `.`, `_`).
    - **Cleaning Phone Numbers**: 
      - Missing values were replaced with empty strings.
      - Non-alphanumeric characters were removed, and phone numbers were reformatted to a standard format (e.g., `123-456-7890`).
      - Invalid placeholders (e.g., `nan--`, `Na--`, `--`) were cleaned up.

4. **Address Parsing**:
    - The `Address` column was split into three separate columns: `Street_adress`, `State`, and `Zip_code`.
    - The original `Address` column was dropped, and missing values were filled with empty strings.

5. **Standardizing Boolean Columns**:
    - The `Do_Not_Contact` and `Paying Customer` columns were standardized to use `Y` (Yes) and `N` (No) for consistency.

6. **Current State of the DataFrame**:
    - The cleaned and transformed DataFrame contains the following columns:
      - `CustomerID`: Unique identifier for each customer.
      - `First_Name` and `Last_Name`: Customer's first and last names.
      - `Phone_Number`: Standardized phone numbers.
      - `Paying Customer` and `Do_Not_Contact`: Boolean indicators for customer preferences.
      - `Street_adress`, `State`, and `Zip_code`: Parsed address components.


```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates().copy()
```

```python
df = df.drop(columns = ['Not_Useful_Column'])
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/._')
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].fillna('').str.replace('[^a-zA-Z0-9]', '', regex=True)

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')
df
```

```python
df[['Street_adress', 'State', 'Zip_code']] = df['Address'].str.split(',', expand=True)
df
```

```python
df.drop(columns=['Address'], inplace=True)
df.fillna('', inplace=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'N')
df
```
